```python
data = pd.read_csv('population.csv')

data
```

```python
data['species'] = data['species'].replace({
    1 : 'species1',
    2 : 'species2',
    3 : 'species3',
    4 : 'species4',
    5 : 'species5',
    6 : 'species6',
    7 : 'species7',
    8 : 'species8',
    9 : 'species9',
    10 : 'species10',
    11 : 'species11',
    12 : 'species12',
    13 : 'species13',
})
```

```python
locs = list(data['location'].unique())
```

```python
dataModel = pd.DataFrame(columns = ['date', 'time', 'boat1', 'boat2', 'boat3', 'boat4', 'boat5', 'boat6', 'boat7', 'boat8', 'boat9', 'boat10'])
```

```python
init = dt.date(2018, 1, 1)
end = dt.date(2022, 1, 1)
delta = dt.timedelta(days= 1)

while init <= end:
    for time in range(1, 5):
        locs = data[data['time'] == time]['location'].unique()
        catches = []
        for boat in range(10):
            location = random.choice(locs)
            temp = data[(data['location'] == location) & (data['time'] == time)]

            dictVals = {'location' : location}

            for val in temp['species']:
                spec = temp[temp['species'] == val]
                size = random.randint(0, spec['size'].values[0])
                dictVals[val] = size
            catches.append(dictVals)
        
        
        dataModel.loc[len(dataModel)] = [init, time] + catches
    init += delta
    
```

```python
historicalData = pd.DataFrame(columns = ['date', 'time', 'species', 'catch', 'effort', 'boat', 'location'])

boats = ['boat1', 'boat2', 'boat3', 'boat4', 'boat5', 'boat6', 'boat7', 'boat8', 'boat9', 'boat10']

for i in range(len(dataModel)):
    row = dataModel.iloc[i]
    date = row['date']
    time = row['time']



    for boat in boats:
        obj = row[boat]
        loc = obj['location']

        species = list(obj.keys())[1:]

        effort = 0
        for spec in species:
            effort += obj[spec]

        for spec in species:   
            historicalData.loc[len(historicalData)] = [date, time, spec, obj[spec], effort, boat,loc]
```
